### (3a) (5 điểm): Tính điểm trung bình

Trong Spark, một pair RDD là một RDD trong đó mỗi phần tử là một tuple `(k, v)` trong đó `k` là key và `v` là value. Cho một pair RDD chứa toàn bộ điểm Pretest của lớp DS, với mỗi pair là tên sinh viên và điểm thi Pretest (xem ví dụ trong phần Test). Sử dụng các hàm map() và reduce() của, viết một function để tính điểm Pretest trung bình của lớp.

```
def averageScore(inputRDD):
  # Your code goes here
  output = inputRDD.mapValues(lambda x: sum(x)).collect()
  return output
```

### (3b) (5 điểm): Sử dụng pair RDD để đếm từ
Sử dụng hàm map() và reduceByKey() của PySpark, viết một function trong PySpark để đếm số lần xuất hiện của mỗi từ trong một RDD. Hàm nhận một RDD đầu vào và trả về một pair RDD

```
def createPairRDD(inputRDD):
  outputRDD = inputRDD.map(lambda x: (x, 1))
  outputRDD.collect()
  
  return outputRDD
```

```
def wordCount(inputRDD):
  RDD = createPairRDD(inputRDD)
  wordsGrouped = RDD.groupByKey()
  wordCountsGrouped = wordsGrouped.mapValues(lambda y: sum(y))
  return wordCountsGrouped
```

```
# Test
inputRDD = sc.parallelize(['cat', 'elephant', 'rat', 'rat', 'cat'])
wordCount(inputRDD).collect()
assert sorted(wordCount(inputRDD).collect()) == [('cat', 2), ('elephant', 1), ('rat', 2)],\
                  'incorrect value for wordCount()'
```

## Bài 5 (30 points)

Trong bài này, bạn sẽ sử dụng mô hình logistic regression để giải quyết bài toán phân lớp nhị phân (binary classification) cho tập dữ liệu ung thư từ thư viện scikit-learn. Sử dụng kĩ thuật chọc hút tế bào bằng kim nhỏ (Fine Needle Aspiration - FNA), hình ảnh khối u vú được sử dụng để tính 30 features trong tập dữ liệu này. Bạn có thể xem thêm chi tiết tại https://scikit-learn.org/stable/datasets/index.html#breast-cancer-dataset. Có hai loại ung thư: malignant (ác tính) và benign (lành tính). Bạn có thể load dữ liệu như dưới đây.

```
# Import scikit-learn dataset library
from sklearn import datasets
# Load dataset
cancer = datasets.load_breast_cancer()
# The feature names and label names can be printed as follows
# print the names of the 13 features
print("Features: ", cancer.feature_names)
# print the label type of cancer('malignant' 'benign')
print("Labels: ", cancer.target_names) 
# The feature data and label data are in \emph{cancer.data} and \emph{cancer.target}.
```

### (5a) (5 điểm). Chia dataset thành training set và test set sử dụng tỉ lệ training set : test set = 7 : 3. Bạn có thể sử dụng hàm sklearn.model_selection.train_test_split().

```
X, y = cancer['data'], cancer['target']
X.shape, y.shape
```

```
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3)
```

### (5b) (5 điểm). Sử dụng hàm sklearn.preprocessing.StandardScaler(), chuẩn hoá tất cả các features của dataset.

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.transform(Xtest)
```

### (5c) (10 điểm). Coi các ví dụ benign là negative class các ví dụ malignant là positive class. Sử dụng mô hình Logistic Regression của scikit-learn linear_model.LogisticRegression và training set để huấn luyện mô hình và kiểm thử trên test set.

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(Xtrain, ytrain)
```

```
print("Accuracy Test", model.score(Xtest, ytest))
```

### (5d) (10 điểm). Tính True Positive rate, False Positive rate, True Negative rate, False Negative rate, Precision, Recall, F1 score, và AUC trên training set và test set. So sánh AUC trên training set and test set.

```
from pprint import pprint
from sklearn.metrics import roc_auc_score, confusion_matrix

ypred = model.predict(Xtest)

tn, fp, fn, tp = confusion_matrix(ytest, ypred).ravel()

tpr = tp / (tp + fn)
fpr = fp / (fp + tn)
fnr = fn / (tp + fn)
tpr = tn / (fp + tn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score =  2 * (precision * recall) / (precision + recall)
auc = roc_auc_score(ytest, ypred)

metrics = {
  'TPR': tpr,
  'FPR': fpr,
  'FNR': fnr,
  'TPR': tpr,
  'Precision': precision,
  'Recall': recall,
  'F1 Score': f1_score,
  'AUC Score': auc
}
pprint(metrics)
```
